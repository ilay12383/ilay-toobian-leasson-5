```python
secret_number = 7
guesses = 0

while True:
    guess = int(input("Guess a number between 1 and 10: "))
    guesses += 1

    if guess > secret_number:
        print("Too high")
    elif guess < secret_number:
        print("Too low")
    else:
        print("Correct!")
        print("You guessed it in", guesses, "guesses")
        break
```

```python
while True:
    print("Menu:")
    print("1 - Add")
    print("2 - Subtract")
    print("3 - Multiply")
    print("4 - Divide")
    print("5 - Exit")

    choice = input("Choose an option (1-5): ")

    if choice == "5":
        break

    if choice in ["1", "2", "3", "4"]:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == "1":
            print(num1 + num2)
        elif choice == "2":
            print(num1 - num2)
        elif choice == "3":
            print(num1 * num2)
        elif choice == "4":
            if num2 == 0:
                print("Cannot divide by zero")
            else:
                print(num1 / num2)
    else:
        print("Invalid choice")
```

```python
even_numbers = []

print("Enter even numbers only (0 to stop)")

while True:
    number = int(input("Enter a number: "))
    
    if number == 0:
        break
    
    if number % 2 != 0:
        print("Only even numbers!")
        continue
    
    even_numbers.append(number)

print("Even numbers entered:", even_numbers)
```

```python
balance = 1000

while True:
    print("\nATM Menu:")
    print("1 - Check balance")
    print("2 - Deposit money")
    print("3 - Withdraw money")
    print("4 - Exit")

    choice = input("Choose an option (1-4): ")

    if choice == "1":
        print("Your balance is:", balance)

    elif choice == "2":
        amount = float(input("Enter amount to deposit: "))
        if amount < 0:
            print("Invalid amount!")
        else:
            balance += amount
            print("Money deposited. New balance:", balance)

    elif choice == "3":
        amount = float(input("Enter amount to withdraw: "))
        if amount < 0:
            print("Invalid amount!")
        elif amount > balance:
            print("Not enough balance!")
        else:
            balance -= amount
            print("Money withdrawn. New balance:", balance)

    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid choice, try again.")

```
