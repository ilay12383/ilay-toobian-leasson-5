```python
numbers = [10, 20, 30, 40, 50]

total = 0
for num in numbers:
    total += num

average = total / len(numbers)

print("Sum:", total)
print("Average:", average)
```

```python
numbers = [23, 67, 12, 89, 45, 34]

maximum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num

print(maximum)
```
